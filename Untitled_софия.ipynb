World Top Resturants

Источник: Kaggle - https://www.kaggle.com/datasets/shaistashahid/world-top-resturants

Бизнес-задачи:

1)В каких странах находится больше всего топовых ресторанов?
2)Какие кухни наиболее популярны среди них?
3)Какие ниболее частые оценки?
4)У каких стран самый высокий средний рейтинг?
5)Как распределены рестораны по уровням цен?
6)Как количество типов кухни влияет на рейтинг?
7)Как распределяются рейтинги ресторанов?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

загрузка файла

```python
df = pd.read_csv('world_top_restaurants_dataset.csv')
```

Удаление дубликатов

```python
df.drop_duplicates(inplace=True)
```

Обработка пропущенных значений (NaN)

```python
df['Cuisine_Type'] = df['Cuisine_Type'].fillna('Unknown')
df['Price_Category'] = df['Price_Category'].fillna('Unknown')
```

Удаление строки без рейтингв

```python
df.dropna(subset=['Rating'], inplace=True)
```

Нормализация и преобразование типов

```python
df['Country'] = df['Country'].str.title()
df['Rating'] = df['Rating'].astype(float)
df['Average_Price_USD'] = df['Average_Price_USD'].astype(float).fillna(0)
```

```python
print(f"Итоговый размер данных после очистки: {df.shape}")
```

Количество типов кухни

```python
df['Cuisine_Count'] = df['Cuisine_Type'].apply(lambda x: len(x.split(',')) if x != 'Unknown' else 0)
```

Уровни цен (usd)

```python
def categorize_by_usd(price):
    if price == 0:
        return 'Unknown'
    elif price < 150:
        return 'Low'
    elif 150 <= price < 300:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Price_Level'] = df['Average_Price_USD'].apply(categorize_by_usd)
```

Уровень рейтинга

```python
def rate_bucket(rating):
    if rating < 4.4:
        return 'Low'
    elif 4.4 <= rating < 4.7:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Rating_Level'] = df['Rating'].apply(rate_bucket)
```

```python
print(f"Уникальные категории цен: {df['Price_Level'].unique()}")
```

ГРафики

```python
sns.set_style("whitegrid")
plt.rcParams['font.size'] = 12
```

№1.  Количество ресторанов по странам

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Country', 
              order=df['Country'].value_counts().head(10).index, 
              palette='Spectral')
plt.title('Топ-10 стран по количеству ресторанов', fontsize=16)
plt.show()
```

№2. Доля ресторанов по уровню цен

```python
price_counts = df['Price_Level'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(price_counts.values, 
        labels=price_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=plt.cm.Set3.colors, 
        wedgeprops={'edgecolor': 'black'})
plt.title('Доля ресторанов по уровню цен', fontsize=16)
plt.axis('equal')
plt.show()
```

№3. Распределение оценок

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Rating'], bins=15, kde=True, color='skyblue')
plt.title('Распределение частоты оценок (Рейтинг)', fontsize=16)
plt.xlabel('Рейтинг')
plt.ylabel('Количество ресторанов')
plt.show()
```

№4. Средний рейтинг в Топ-10 странах

```python
top_countries = df['Country'].value_counts().head(10).index
avg_rating_country = df[df['Country'].isin(top_countries)].groupby('Country')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
sns.barplot(x=avg_rating_country.index, y=avg_rating_country.values, palette='viridis')
plt.title('Средний рейтинг ресторанов в Топ-10 странах', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.show()
```

№5. Влияние количества типов кухни на рейтингш

```python
plt.figure(figsize=(8, 5))
sns.countplot(
    data=df, 
    x='Price_Level', 
    order=['Low', 'Medium', 'High', 'Unknown'], 
    palette='coolwarm'
)
plt.title('Распределение ресторанов по уровню цен', fontsize=16)
plt.xlabel('Уровень цены')
plt.ylabel('Количество ресторанов')
plt.show()
```

№6.  Распределение рейтинга по ценам

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='Price_Level', y='Rating', 
               order=['Low', 'Medium', 'High', 'Unknown'], palette='pastel')
plt.title('Сравнение распределения рейтинга по уровню цен', fontsize=16)
plt.xlabel('Уровень цены')
plt.ylabel('Рейтинг')
plt.show()
```

№7. Самые популярные кухни

```python
cuisine_list = df['Cuisine_Type'].str.split(',').explode().str.strip()
plt.figure(figsize=(10, 6))
sns.countplot(y=cuisine_list, 
              order=cuisine_list.value_counts().head(10).index, 
              palette='rocket')
plt.title('Топ-10 самых распространенных типов кухни', fontsize=16)
plt.show()
```

№8. Связь рейтинга и цены

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Average_Price_USD', y='Rating', 
                hue='Price_Level', palette='tab10', alpha=0.7)
plt.title('Зависимость рейтинга от средней цены (USD)', fontsize=16)
plt.xlabel('Средняя цена (USD)')
plt.ylabel('Рейтинг')
plt.xlim(0, df['Average_Price_USD'].quantile(0.95))
plt.legend(title='Уровень цены')
plt.show()
```

№9. Распределение по годам основания

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(df['Established_Year'].dropna().astype(int), fill=True, color='purple', linewidth=3)
plt.title('Плотность распределения ресторанов по году основания', fontsize=16)
plt.xlabel('Год основания')
plt.ylabel('Плотность')
plt.show()
```

№10. Распределение ресторанов по континентам

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Continent', y='Rating', errorbar=None, palette='mako')
plt.title('Средний рейтинг ресторанов по континентам', fontsize=16)
plt.xlabel('Континент')
plt.ylabel('Средний рейтинг')
plt.show()
```
